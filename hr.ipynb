## Data cleaning:
Data cleaning helps to eliminate inconsistencies, handle missing values, and remove duplicates, improving the quality and integrity of your dataset, ultimately leading to more accurate analysis and reliable insights.

```python
missing_values=data.isnull().sum() #returns the count of missing values for each column in the DataFrame 'data'    
print(missing_values)
```

```python
no_of_duplicates=data.duplicated().sum() #total no of duplicate rows from the datadrame
print(no_of_duplicates)
```

```python
data=data.drop_duplicates() #removes duplicate rows from the DataFrame and assigns the cleaned DataFrame back to the variable 'data'
data
```

```python
data.shape #returns a tuple representing the dimensions (rows, columns) of the DataFrame 'data'
```

```python
data.isnull().sample(10)
```

```python
# Fill missing values in 'YearsWithCurrManager' with the mode value
mode_value = data['YearsWithCurrManager'].mode()
print(mode_value)
data['YearsWithCurrManager'].fillna(mode_value, inplace=True)

# Verify if missing values have been filled
missing_values_after = data['YearsWithCurrManager'].isnull().sum()
print(f'Missing values after filling: {missing_values_after}')
```

```python
data.isnull().sum()
```

## Data Visualization

```python
# Age Distribution
plt.figure(figsize=(10, 6))
sns.histplot(data['Age'], bins=15, kde=True)
plt.title('Age Distribution of Employees')
plt.xlabel('Age')
plt.ylabel('Number of Employees')
plt.show()
```

```python
# Gender Distribution
plt.figure(figsize=(10, 6))
sns.countplot(x='Gender', data=data)
plt.title('Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('Number of Employees')
plt.show()
male_count = data[data['Gender'] == 'Male'].shape[0]
female_count = data[data['Gender'] == 'Female'].shape[0]
print('male count=',male_count,'famale count = ',female_count)
```
